```python
import os
import pandas as pd 
import numpy as np
from sklearn.ensemble import RandomForestClassifier

train_dir = '/kaggle/input/titanic/train.csv'
test_dir = '/kaggle/input/titanic/test.csv'

def read_dataset(ds):
    data = pd.read_csv(ds)
    return data

datatrain = read_dataset(train_dir)
datatest = read_dataset(test_dir)
```

```python
ytrain = datatrain["Survived"]

resources = ["Sex","SibSp", "Pclass", "Parch"]
xtest = pd.get_dummies(datatest[resources])
xtrain = pd.get_dummies(datatrain[resources])


graph = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
graph.fit(xtrain, ytrain)


ytest = graph.predict(xtest)
answer = pd.DataFrame({'PassengerId': datatest.PassengerId, 'Survived': ytest})
answer.to_csv('submission.csv', index=False)
```
